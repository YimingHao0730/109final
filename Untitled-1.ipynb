```python
import pandas as pd
```

```python
df = pd.read_csv('train.csv')
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split


# Separate features and target variable
X = df.drop("SalePrice", axis=1)
y = df["SalePrice"]

# Preprocess the data
numeric_cols = X.select_dtypes(include=["int64", "float64"]).columns
categorical_cols = X.select_dtypes(include=["object"]).columns

# Handle missing values
numeric_imputer = SimpleImputer(strategy="mean")
X[numeric_cols] = numeric_imputer.fit_transform(X[numeric_cols])

categorical_imputer = SimpleImputer(strategy="most_frequent")
X[categorical_cols] = categorical_imputer.fit_transform(X[categorical_cols])

# Encode categorical variables using one-hot encoding
one_hot_encoder = OneHotEncoder(drop="first", sparse=False, handle_unknown="ignore")
X_encoded = pd.DataFrame(one_hot_encoder.fit_transform(X[categorical_cols]))
X_encoded.columns = one_hot_encoder.get_feature_names(categorical_cols)

# Concatenate encoded features with numeric features
X_preprocessed = pd.concat([X[numeric_cols], X_encoded], axis=1)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

# Model 1: Linear Regression
linear_model = LinearRegression()
linear_scores = cross_val_score(linear_model, X_train, y_train, cv=10, scoring="neg_mean_squared_error")
linear_rmse = np.sqrt(-linear_scores.mean())

# Model 2: Gradient Boosting Regressor
gb_model = GradientBoostingRegressor()
gb_scores = cross_val_score(gb_model, X_train, y_train, cv=10, scoring="neg_mean_squared_error")
gb_rmse = np.sqrt(-gb_scores.mean())

# Model 3: K-Nearest Neighbors Regressor
knn_model = KNeighborsRegressor(n_neighbors=5)
knn_scores = cross_val_score(knn_model, X_train, y_train, cv=10, scoring="neg_mean_squared_error")
knn_rmse = np.sqrt(-knn_scores.mean())

# Plotting the model comparison
models = ['Linear Regression', 'Gradient Boosting', 'KNN']
rmse_scores = [linear_rmse, gb_rmse, knn_rmse]

plt.bar(models, rmse_scores)
plt.xlabel('Models')
plt.ylabel('RMSE')
plt.title('Model Comparison - RMSE')
plt.show()

# Print the results
print("Linear Regression RMSE:", linear_rmse)
print("Gradient Boosting Regressor RMSE:", gb_rmse)
print("K-Nearest Neighbors Regressor RMSE:", knn_rmse)
```

```python
from sklearn.metrics import mean_squared_error

# Train the Gradient Boosting Regressor on the full training dataset
gb_model.fit(X_train, y_train)

# Make predictions on the testing set
y_pred = gb_model.predict(X_test)

# Calculate the root mean squared error (RMSE) of the predictions
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# Visualize the predictions and actual values
plt.figure(figsize=(10, 6))
plt.scatter(range(len(y_test)), y_test, color='blue', label='Actual')
plt.scatter(range(len(y_pred)), y_pred, color='red', label='Predicted')
plt.title('Actual vs. Predicted Sales Price')
plt.xlabel('Index')
plt.ylabel('Price')
plt.legend()
plt.show()

# Print the final parameter estimates and RMSE
print("Final Parameter Estimates:")
print(gb_model.get_params())
print("RMSE:", rmse)
```

```python
r2_score = gb_model.score(X_test, y_test)
print("R^2 Score:", r2_score)
```
